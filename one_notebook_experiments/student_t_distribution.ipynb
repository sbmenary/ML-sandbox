```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
true_mean   = 10
true_std    = 2
num_samples = 6
num_toys    = int(1e5)
```

```python
data    = true_mean + true_std*np.random.normal(size=(num_toys, num_samples))
data_mu = np.mean(data, axis=1)
data_S  = np.std (data, axis=1) *  np.sqrt(num_samples) / np.sqrt(num_samples - 1)
data_X  = (data_mu - true_mean) / (data_S / np.sqrt(num_samples))
```

```python
fig = plt.figure(figsize=(5, 5))
ax  = fig.add_subplot(1, 1, 1)
hvals, bins, _ = ax.hist(data_X, bins=100, density=True)
ax.plot(bins, stats.t.pdf(bins, df=num_samples-1), "--", c="r", lw=2)
plt.show(fig)
plt.close(fig)
```
